```python
# LN, DLQ, CD에서 YM이 201805, 201806, 201807, 201808, 201809, 201810, 201811이 아닌 것은 모두 없애기

LN=LN[LN["YM"]>201804]
DLQ=DLQ[DLQ["YM"]>201804]
CD=CD[CD["YM"]>201804]
```

```python
# ID와 LN을 JOIN_KEY를 KEY로 하여 MERGE

NEW_LN = pd.merge(ID, LN, how='outer', on = 'JOIN_KEY')
NEW_LN
```

```python
# 결측값(ym)이 들어있는 행 전체 삭제

NEW_LN = NEW_LN.dropna(axis=0)
NEW_LN
```

```python
NEW_LN = NEW_LN[NEW_LN['YM'].notna()]
NEW_LN['YM'] = NEW_LN['YM'].astype(int)
NEW_LN
```

```python
NEW_LN.info()
```

```python
NEW_LN=NEW_LN.astype({'YM':'object'})
```

```python
# JOIN_KEY와 YM을 합쳐야한다. 둘다 KEY값이다

cols = ['YM','JOIN_KEY']
NEW_LN['JOIN_KEY_YM'] = NEW_LN[cols].apply(lambda row:'_'.join(row.values.astype(str)),
                            axis=1)
DLQ['JOIN_KEY_YM'] = DLQ[cols].apply(lambda row:'_'.join(row.values.astype(str)),
                            axis=1)
CD['JOIN_KEY_YM'] = CD[cols].apply(lambda row:'_'.join(row.values.astype(str)),
                            axis=1)
```

```python
# JOIN 했으니까 JOIN_KEY와 YM은 버린다

NEW_LN = NEW_LN.drop('JOIN_KEY',axis=1)
NEW_LN = NEW_LN.drop('YM',axis=1)
DLQ = DLQ.drop('JOIN_KEY',axis=1)
DLQ = DLQ.drop('YM',axis=1)
CD = CD.drop('JOIN_KEY',axis=1)
CD = CD.drop('YM',axis=1)
```

```python
# DLQ에 있는 SCTR_CD, COM_KEY, DLQ_TYPE, DLQ_CD_1, DLQ_CD_2, DLQ_YM을 모두 삭제한다

DLQ = DLQ.drop('SCTR_CD',axis=1)
DLQ = DLQ.drop('COM_KEY',axis=1)
DLQ = DLQ.drop('DLQ_CD_1',axis=1)
DLQ = DLQ.drop('DLQ_CD_2',axis=1)
DLQ = DLQ.drop('DLQ_YM',axis=1)
print(DLQ.head())
```

```python
# DLQ에 있는 DLQ_SCTR_CD을 모두 1로 채우기

DLQ['SCTR_CD'] = '1'
DLQ
```

```python
NEW_LN
```

```python
NEW_LN['NUM_LN'] = NEW_LN.groupby(['JOIN_KEY_YM'])['LN_AMT'].transform('count')
NEW_LN
```

```python
NEW_LN['SUM_LN'] = NEW_LN.groupby('JOIN_KEY_YM')['LN_AMT'].transform('sum')
NEW_LN
```

```python
NEW_LN['AVE_LN'] = NEW_LN['SUM_LN'] / NEW_LN['NUM_LN']
NEW_LN
```

```python
NEW_LN = NEW_LN.loc[NEW_LN.groupby('JOIN_KEY_YM')['LN_YM'].idxmin()]
NEW_LN
```

```python
NEW_LN_DLQ = pd.merge(NEW_LN, DLQ, on='JOIN_KEY_YM', how='outer')
NEW_LN_DLQ
```

```python
X = NEW_LN_DLQ[['BTH_YR','GENDER','SCTR_CD_x','LN_CD_1','LN_CD_2','LN_YM','LN_AMT','NUM_LN','SUM_LN','AVE_LN']].values
y = NEW_LN_DLQ[['DLQ_TYPE','DLQ_AMT','SCTR_CD_y']].values
X
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(f'{device} is available')
```

```python
def seq_data(x, y, sequence_length):

  x_seq = []
  y_seq = []
  for i in range(len(x) - sequence_length):
    x_seq.append(x[i: i+sequence_length])
    y_seq.append(y[i+sequence_length])

  return torch.FloatTensor(x_seq).to(device), torch.FloatTensor(y_seq).to(device).view([-1, 1])
```

```python
seq_data(X, y, 5)
```

```python
split = 200
sequence_length = 5

x_seq, y_seq = seq_data(X, y, sequence_length)

x_train_seq = x_seq[:split]
y_train_seq = y_seq[:split]
x_test_seq = x_seq[split:]
y_test_seq = y_seq[split:]
print(x_train_seq.size(), y_train_seq.size())
print(x_test_seq.size(), y_test_seq.size())
```

```python
train = torch.utils.data.TensorDataset(x_train_seq, y_train_seq)
test = torch.utils.data.TensorDataset(x_test_seq, y_test_seq)

batch_size = 20
train_loader = torch.utils.data.DataLoader(dataset=train, batch_size=batch_size, shuffle=False)
test_loader = torch.utils.data.DataLoader(dataset=test, batch_size=batch_size, shuffle=False)
```
